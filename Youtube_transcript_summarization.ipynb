```python
import re
```

```python
symbols_to_remove = "#&*()[]{};:,<>/|\\\""
```

```python
def preprocess(text):
        text=re.sub(r'\s+', ' ', text).strip()
        text=re.sub(r'\(.*?\)', '', text).strip()
        text=re.sub(r'\[.*?\]', '', text).strip()
        text=re.sub(r'\w*\d\w*', '', text).strip()
        text=re.sub(f"[{re.escape(symbols_to_remove)}]", ' ', text)
        return text
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('words')
```

```python
english_words = set(nltk.corpus.words.words())
```

```python
def split_joint_words(text):    
    words = text.split()
    split_text = []
    for i in words:
        if i in english_words:
            split_text.append(i)
            continue
        split_found = False
        for j in range(2,len(i)-2):
            left = i[:j]
            right = i[j:]
            if left in english_words and right in english_words:
                split_text.append(left)
                split_text.append(right)
                split_found = True
                break
      
        if not split_found:
            split_text.append(i)
    
    return ' '.join(split_text)
```

```python
df_final['transcript']=df_final['transcript'].astype(str).apply(split_joint_words)
```

```python
filler_words = ["um", "uh","ah","err","hmm","ehh","you know", "like", "actually", "basically", "so", "well", "literally"]
```

```python
def remove_filler_words(text):
    for filler in filler_words:
        text = re.sub(r'\b' + re.escape(filler) + r'\b', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df_final['transcript']=df_final['transcript'].astype(str).apply(remove_filler_words)
```

```python
def remove_stopwords(text):
    new_text=[]
    for word in text.split():
        if word in stopwords_set:
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x).strip()
        
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def compute_tfidf(sentences):
    sentence_list = [" ".join(sentence) for sentence in sentences]
    vectorizer = TfidfVectorizer()
    try:
        tfidf_matrix = vectorizer.fit_transform(sentence_list)
        return tfidf_matrix.toarray().sum(axis=1)  
    except ValueError as e:
        return []
```

```python
df_final2['tfidf_scores'] = df_final2['lemmatized_transcript'].apply(lambda sentences: compute_tfidf(sentences))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def cohesion_score(sentences):
    cohesion_scores = []
    for i in range(1, len(sentences)):
        prev_sentence = " ".join(sentences[i-1])
        curr_sentence = " ".join(sentences[i])
        vectorizer = TfidfVectorizer()
        vectors = vectorizer.fit_transform([prev_sentence, curr_sentence])
        cohesion = cosine_similarity(vectors[0], vectors[1])[0][0]
        cohesion_scores.append(cohesion)
    return [0] + cohesion_scores  
```

```python
df_final2['cohesion_score'] = df_final2['lemmatized_transcript'].apply(lambda sentences: cohesion_score(sentences))
```

```python
def get_top_5_percent_summary(ranked_sentences, top_percent=5):
    num_sentences = len(ranked_sentences)
    if num_sentences == 0:
        return ""
    
    top_n = max(1, int(num_sentences * top_percent / 100))
   
    top_sentences = [sentence for sentence, _ in ranked_sentences[:top_n]]
    
    return ''.join([''.join(sentence) for sentence in top_sentences])

df_final2['summary'] = df_final2['ranked_sentences'].apply(get_top_5_percent_summary)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from rouge_score import rouge_scorer
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

```

```python
def generate_top_5_percent_summary(text, num_sentences=5):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, num_sentences)
    summary_sentences = [str(sentence) for sentence in summary]
    num_sentences_to_keep = max(1, int(len(summary_sentences) * 0.05))    
    top_sentences = summary_sentences[:num_sentences_to_keep]
    final_summary = ' '.join(top_sentences)
    
    return final_summary

```

```python
def compute_rouge_scores(reference, generated):
    scores = scorer.score(reference, generated)
    return scores
```

```python

df_final2['generated_summary'] = df_final2['transcript'].apply(lambda x: generate_top_5_percent_summary(x, num_sentences=5))
df_final2['rouge_scores'] = df_final2.apply(lambda row: compute_rouge_scores(row['summary'], row['generated_summary']), axis=1)

```
