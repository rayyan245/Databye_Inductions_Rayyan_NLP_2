```python
from bs4 import BeautifulSoup
import requests
import numpy
import pandas as pd
import csv
```

```python
# URL of the website (replace with the actual URL)
url = "https://www.sbicard.com/sbi-card-en/assets/docs/html/personal/offers/reliance-digital-list.html"

# Send a GET request to fetch the HTML content of the webpage
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all <tr> tags (each row of the table)
rows = soup.find_all('tr')
data = []
# Loop through each row to extract the data from <td> tags
for row in rows:
    # Find all <td> tags in the current row
    columns = row.find_all('td')

    # Extract the text from each <td> (strip leading/trailing whitespace)
    if len(columns) >= 5:  # Ensure there are enough columns in the row
        store_name = columns[0].get_text(strip=True)
        city = columns[1].get_text(strip=True)
        pincode = columns[2].get_text(strip=True)
        address_part1 = columns[3].get_text(strip=True)
        address_part2 = columns[4].get_text(strip=True)

        # Combine address parts if necessary
        full_address = address_part1 + " " + address_part2
        data.append([store_name, city, pincode, full_address])

        # Print or store the extracted data
        print(f"Store Name: {store_name}")
        print(f"City: {city}")
        print(f"Pincode: {pincode}")
        print(f"Address: {full_address}")
        print("-" * 50)
```

```python
df=pd.DataFrame(data,columns=['Store Name','City','Pincode','Address'])
```

```python
df.drop(df.index[0], inplace=True)
```

```python
df.drop(df.index[0], inplace=True)
```

```python
df.drop(columns=['Address'],inplace=True)
```

```python
df.to_csv('Reliance_stores_final.csv',index=False)
```

```python

# Specify the path to your PDF file
pdf_file_path = "/content/store-bajaj-oct24.pdf"
# Open the PDF file
with pdfplumber.open(pdf_file_path) as pdf:
    # Initialize an empty list to hold DataFrames
    tables = []

    # Iterate through the pages of the PDF
    for page in pdf.pages:
        # Extract tables from the current page
        page_tables = page.extract_tables()

        # Loop through the extracted tables and convert them to DataFrames
        for table in page_tables:
            # Create a DataFrame from the table
            df = pd.DataFrame(table[1:], columns=table[0])  # First row as header
            tables.append(df)  # Add the DataFrame to the list

# Combine all DataFrames into one (if needed)
#final_df = pd.concat(tables, ignore_index=True)

# Display the final DataFrame
print(tables)
```

```python
table_headers=['S No.','Store Name', 'Address', 'City', 'State', 'Pin code']
```

```python
for table in tables[1:]:
  table_head = pd.DataFrame([table.columns], columns=table.columns)
  table_head.columns=table_headers
  table=pd.DataFrame(table.values)
  table.columns=table_headers
  df=pd.concat([df,table],ignore_index=True)
```
